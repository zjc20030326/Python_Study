```python
from collections import defaultdict

positions = defaultdict(list)

# 示例数据
a = [1, 2, 3, 2, 1, 2, 3, 2]

# 遍历数组，记录每个数字出现的位置（从1开始计数）
for i in range(len(a)):
    positions[a[i]].append(i + 1)

# 输出 positions
for key, value in positions.items():
    print(f"数字 {key} 出现的位置: {' '.join(map(str, value))}")
```

```python
from collections import Counter

T = int(input())

for _ in range(T):
    n = int(input())
    words = input().split()

    common = Counter(words[0])

    for s in words[1:]:
        common &= Counter(s)

    ans = "".join(ch * common[ch] for ch in sorted(common))

    if ans:
        print(ans)
    else:
        print("-1")
```

```python
import bisect

arr = [0, 1, 2, 5, 6]
# 输入查询的元素
x = 3


first = bisect.bisect_left(
    arr, x
)  # bisect_left：查找目标值的第一个不小于目标值的位置。
last = bisect.bisect_right(arr, x)  # bisect_right：查找目标值的第一个大于目标值的位置。
print(first)
print(last)
```
