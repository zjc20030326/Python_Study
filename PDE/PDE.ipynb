```python
import numpy as np
import matplotlib.pyplot as plt

# 1. 定义域离散化：
# 在区间 [0, 2π] 上均匀取 100 个点
# linspace = linear space (线性空间)
x = np.linspace(0, 2 * np.pi, 100)

# 2. 算子作用：
# 计算正弦函数和它的导数 (近似)
y = np.sin(x)
y_derivative = np.cos(x)

# 3. 可视化 (映射到几何)
plt.figure(figsize=(8, 4))  # 创建画布
plt.plot(x, y, label="f(x) = sin(x)", color="blue")
plt.plot(x, y_derivative, label="f'(x) = cos(x)", color="red", linestyle="--")

# 添加数学标注
plt.title("Function vs Derivative")
plt.xlabel("Domain (x)")
plt.ylabel("Value")
plt.legend()  # 显示图例
plt.grid(True)  # 显示网格 (类似于坐标纸)

# 显示图像
plt.show()
```

```python
import os

os.environ["KMP_DUPLICATE_LIB_OK"] = "TRUE"
import torch

# 1. 模拟一个图 (Graph)
# 假设有 3 个节点，每个节点有 4 个特征
# 这里的 X 就是 PDE 里的函数值 u(x)
X = torch.rand(3, 4)

# 2. 模拟邻接矩阵 (Adjacency Matrix)
# 1 表示相连，0 表示不连。这里是一个环状图：1-2, 2-3, 3-1
A = torch.tensor([[1.0, 1.0, 0.0], [0.0, 1.0, 1.0], [1.0, 0.0, 1.0]])

# 3. GCN 的核心操作：聚合 (Aggregation)
# 相当于 PDE 里的拉普拉斯平滑：每个节点的新特征 = 自己 + 邻居
X_new = torch.mm(A, X)  # 矩阵乘法

print("原始特征 X:\n", X)
print("\n聚合后的特征 X_new (一次热扩散):\n", X_new)
```
