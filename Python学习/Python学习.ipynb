# Jupyter Notebook 快捷键指南

## 1. 创建新单元格

- 在当前单元格下方创建: `B`
- 在当前单元格上方创建: `A`

## 2. 删除单元格

- 快速删除: `D D`（连按两次 D）

## 3. 在单元格间跳转

- 跳转到上一个单元格: `Up Arrow`（上方向键）或 `K`
- 跳转到下一个单元格: `Down Arrow`（下方向键）或 `J`

## 4. 运行代码

- 运行当前单元格: `Ctrl + Enter`
- 运行当前单元格并选中下一个: `Shift + Enter`
- 运行当前单元格并在下方插入新单元格: `Alt + Enter`
- 运行当前单元格以上所有单元格: `Ctrl + Alt + P`
- 运行所有单元格: `Ctrl + Alt + R`

## 使用技巧

- 这些快捷键在**命令模式**下有效。单元格呈蓝色边框时为命令模式。
- 按 `Esc` 可以从编辑模式切换到命令模式。
- 按 `Enter` 可以从命令模式切换到编辑模式。
- 运行代码的快捷键在编辑模式下也能使用，非常方便。

```python
while True:
    command = input("请输入命令（exit/pow）：")
    if command == "pow":
        base = float(input("请输入底数："))
        exp = float(input("请输入指数："))
        pow(base, exp)
    elif command == "exit":
        break
```

```python
i = 1
while i <= 9:
    j = 9
    while j >= i:
        print(j, "*", i, "=", i * j, end=" ")
        j = j - 1
    print()
    i = i + 1
```

```python
i = 2
while i < 10:
    isprime = True
    j = 2
    while j * j <= i:
        if i % j == 0:
            isprime = False
            print(i, "=", j, "*", i // j)
            break
        else:
            j += 1
    if isprime:
        print(i, "是一个素数")
    i += 1
```

```python
n = 2
while n < 10:
    x = 2
    while x < n:
        if n % x == 0:
            print(n, "=", x, "*", n // x)  # 整除
            break
        x += 1
    else:
        print(n, "是一个素数")
    n += 1
```
